```python
# importong the redquired modules

from tkinter import *

def plus():
    a = name.get()
    a = a+" + "
    return name.set(a)

def minus():
    a = name.get()
    a = a+" - "
    return name.set(a)

def mul():
    a = name.get()
    a = a+" * "
    return name.set(a)

def divide():
    a = name.get()
    a = a+" / "
    return name.set(a)

def percent():
    a = name.get()
    a = a+" % "
    return name.set(a)

def ac():
    return name.set(" ")

def delete():
    a = name.get()
    a = a[:-1]
    return name.set(a)

def one():
    a = name.get()
    a = a+"1"
    return name.set(a)

def two():
    a = name.get()
    a = a+"2"
    return name.set(a)

def three():
    a = name.get()
    a = a+"3"
    return name.set(a)

def four():
    a = name.get()
    a = a+"4"
    return name.set(a)

def five():
    a = name.get()
    a = a+"5"
    return name.set(a)

def six():
    a = name.get()
    a = a+"6"
    return name.set(a)

def seven():
    a = name.get()
    a = a+"7"
    return name.set(a)

def eight():
    a = name.get()
    a = a+"8"
    return name.set(a)

def nine():
    a = name.get()
    a = a+"9"
    return name.set(a)

def dot():
    a = name.get()
    a = a+"."
    return name.set(a)

def zero():
    a = name.get()
    a = a+"0"
    return name.set(a)

def equal():
    a = name.get()
    b =eval(a)
    return name.set(b)

windows = Tk()

windows.configure(background='black')

name = StringVar()
text = Entry(windows,width = 30, textvariable = name)
text.grid(row=0, column=0, columnspan=4)

ac = Button(windows,text = "AC", width = 5, command = ac)
ac.grid(row=2, column=0)

delete = Button(windows,text = "C", width = 5, command = delete)
delete.grid(row=2, column=1,)

percent = Button(windows,text = "%", width = 5, command = percent)
percent.grid(row=2, column=2)

divide = Button(windows,text = "/", width = 5, command = divide)
divide.grid(row=2, column=3)

seven = Button(windows,text = "7", width = 5, command = seven)
seven.grid(row=3, column=0)

eight = Button(windows,text = "8", width = 5, command = eight)
eight.grid(row=3, column=1,)

nine = Button(windows,text = "9", width = 5, command = nine)
nine.grid(row=3, column=2)

mul = Button(windows,text = "x", width = 5, command = mul)
mul.grid(row=3, column=3)

four = Button(windows,text = "4", width = 5, command = four)
four.grid(row=4, column=0)

five = Button(windows,text = "5", width = 5, command = five)
five.grid(row=4, column=1,)

six = Button(windows,text = "6", width = 5, command = six)
six.grid(row=4, column=2)

minus = Button(windows,text = "-", width = 5, command = minus)
minus.grid(row=4, column=3)

one = Button(windows,text = "1", width = 5, command = one)
one.grid(row=5, column=0)

two = Button(windows,text = "2", width = 5, command = two)
two.grid(row=5, column=1,)

three = Button(windows,text = "3", width = 5, command = three)
three.grid(row=5, column=2)

plus = Button(windows,text = "+", width = 5, command = plus)
plus.grid(row=5, column=3)

dot = Button(windows,text = ".", width = 5, command = dot)
dot.grid(row=6, column=0)

zero = Button(windows,text = "0", width = 5, command = zero)
zero.grid(row=6, column=1,)

equal = Button(windows,text = "=", width = 11, command = equal)
equal.grid(row=6, column=2,columnspan = 2)

windows.mainloop()
```
